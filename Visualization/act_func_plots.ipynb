```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.array([-1, 0, 1])
y = np.array([-0.1, 0, 1])
# plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.plot(x, y, 'k')
plt.savefig('Leaky ReLU')
```

```python
x = np.linspace(-5, 5, 100)
y = 1 / (1 + pow(math.e, -x))
# plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.plot(x, y, 'k')
plt.savefig('sigmoid')
```

```python
x = np.array([-1, 0, 1])
y = np.array([0, 0, 1])
# plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.step(x, y, 'k')
plt.savefig('Step Func')
```
