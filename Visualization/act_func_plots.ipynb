```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.array([-1, 0, 1])
y = np.array([-0.1, 0, 1])
# plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.plot(x, y, 'k')
# plt.savefig('Leaky ReLU')
```

```python
x = np.linspace(-5, 5, 100)
y = 1 / (1 + pow(math.e, -x))
# plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.plot(x, y, 'k')
# plt.savefig('sigmoid')
```

```python
x = np.array([-1, 0, 1])
y = np.array([0, 0, 1])
# plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.step(x, y, 'k')
# plt.savefig('Step Func')
```

```python
x = np.linspace(-5, 5, 100)
y = pow(x, 2)
# plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.plot(x, y, 'k')
plt.savefig('bell curve')
```

```python
# import for colormaps
from matplotlib import cm

x=np.linspace(-10, 10, num=100)
y=np.linspace(-10, 10, num=100)

x, y = np.meshgrid(x, y)

z = (x**2)+(y**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d',)
ax.plot_surface(x,y,z, cmap=cm.jet)
# plt.show()
plt.savefig('3D inverse bell', dpi=600)
```
